```python
# imports
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm2
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
from sklearn.model_selection  import train_test_split
import numpy as np

# allow plots to appear directly in the notebook
%matplotlib inline
```

```python
df_macro = pd.read_csv('Macro_Economic.csv', header=0)
```

```python
df_macro.head()
```

```python
df_macro["YearQTR"] = df_macro["year"].astype(str) + '-'+ df_macro["quarter"].astype(str)
```

```python
df_macro.set_index('YearQTR')
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_macro, df_macro['infl'],test_size=0.3, random_state=0)
```

```python
x_train.set_index('YearQTR')
```

```python
x_test.set_index('YearQTR')
```

```python
lm1 = smf.ols(formula="infl ~ tbilrate", data=x_train).fit()
lm1.summary()
```

```python
lm1 = smf.ols(formula="infl ~ tbilrate -1", data=x_train).fit()
lm1.summary()
```

```python
lm2 = smf.ols(formula="infl ~ tbilrate + realint -1", data=x_train).fit()
lm2.summary()
```

```python
lm2_predict=lm2.predict(x_test)
```

```python
lm2_predict.describe()
```

```python
predict_values=pd.concat([x_test["infl"], x_test['YearQTR'],lm2_predict],axis=1).set_index('YearQTR')
predict_values.columns=["actual_inflation","predicted_inflation"]
predict_values["residual"]=predict_values["actual_inflation"]-predict_values["predicted_inflation"]
```

```python
predict_values.head()
```

```python
mae=metrics.mean_absolute_error(predict_values['actual_inflation'], predict_values['predicted_inflation'])
mse=metrics.mean_squared_error(predict_values['actual_inflation'], predict_values['predicted_inflation'])
rmse=np.sqrt(metrics.mean_squared_error(predict_values['actual_inflation'], predict_values['predicted_inflation']))
```

```python
print ('mean absolute error', mae)
print ('mean squared error', mse)
print ('root mean squared error', rmse)
```
